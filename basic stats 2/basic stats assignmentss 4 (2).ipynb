```python
patient_dfs=patient_df[['PatientID', 'Department', 'Doctor', 'BillAmount']] #Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].
```

```python
patient_dfs
```

```python
patient_dfs.drop(columns=["ReceptionistID","CheckInTime"],errors="ignore") #Drop administrative columns like ['ReceptionistID', 'CheckInTime'].
```

```python
depts_bills=patient_dfs.groupby("Department")["BillAmount"].sum().reset_index() #Use groupby to find total bill amount per department.
```

```python
depts_bills
```

```python
drop_dupls=patient_dfs.drop_duplicates("PatientID") #Remove duplicate patient records based on PatientID.
```

```python
drop_dupls
```

```python
patient_dfs.fillna(patient_dfs["BillAmount"].mean(),inplace=True)#Fill missing BillAmount values with the mean bill amount.
```

```python
patient_dfs
```

```python
billing_data=pd.read_csv(r"C:\Users\USER\Downloads\Billing_Data.csv")
```

```python
billing_data
```

```python
comb_data=pd.merge(patient_dfs,billing_data,on="PatientID",how="left") #Merge the billing dataset with patient dataset on PatientID.
```

```python
comb_data
```

```python
new_patients = pd.DataFrame({
    'PatientID': [2001, 2002],
    'Department': ['Cardiology', 'Neurology'],
    'Doctor': ['Dr. Rao', 'Dr. Sharma'],
    'BillAmount': [12500, 9800],
    "InsuranceCovered":[50000,58000],
    "FinalAmount":[55000,60000]
})
```

```python
final_df = pd.concat([comb_data, new_patients], axis=0, ignore_index=True) #Concatenate an additional DataFrame that contains new patients for the current week (row-wise).
```

```python
final_df
```

```python
new_billing_categories = pd.DataFrame({
    'PatientID': [101, 102,103,104,105,106,107,112],
    'InsuranceCovered': [True, False,True,False,True,False,True,True],
    'FinalAmount': [2000, 3000,2500,3500,5000,5500,6000,12000]
})
```

```python
new_billing_categories
```

```python
finals_data = pd.concat([final_df, new_billing_categories], axis=1)#Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).
```
