```python
def outlier_capping(df,column):
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    iqr=q3-q1
    lower_extream=q1-1.5*iqr
    upper_extream=q3+1.5*iqr
    df[column]=df[column].apply(lambda x : lower_extream if x<lower_extream else upper_extream if x>upper_extream else x)
for col in df.select_dtypes(["int","float"]).columns:
    outlier_capping(df,col)        
```

1.Min-Max Scaler (Normalization)
 Range: Scales features to a fixed range [0, 1] (or any range you specify).
 Effect: Shrinks all values proportionally so the minimum becomes 0 and the maximum becomes 1.
 Best for: When you want all features to have the same scale but preserve the shape of the original distribution.
 Sensitive to outliers (because min and max values are affected by extreme values).

Used in algorithms that depend on distance â€” e.g.:
KNN
K-Means
Neural Networks

2.Standard Scaler (Z-score Normalization)
 mean of the feature
 standard deviation
 Range: No fixed range â€” data is centered around 0 with a standard deviation of 1.
 Effect: Removes the mean and scales to unit variance.
 Less affected by outliers (compared to MinMax).

Used in algorithms that assume normally distributed data or use weights â€” e.g.:
Linear Regression
Logistic Regression
SVM
PCA

```python
df['age_group'] = pd.cut(df['age'],bins=[-2, -0.5, 0.5, 1.5],labels=['Young', 'Adult', 'Senior'])
```

```python
df['capital_net'] = df['capital_gain'] - df['capital_loss']
```
