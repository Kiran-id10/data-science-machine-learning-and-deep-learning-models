```python
def outlier_capping(df,column):
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    iqr=q3-q1
    lower_extream=q1-1.5*iqr
    upper_extream=q3+1.5*iqr
    df[column]=df[column].apply(lambda x:lower_extream if x<lower_extream else upper_extream if x>upper_extream else x)
for col in df.select_dtypes(["int","float"]).columns:
    outlier_capping(df,col)
```

1.Clinical / Risk Assessment
The increase in baseline heart rate (LB) and short-term variability (ASTV) may indicate rising fetal activity or mild stress.
The decrease in long-term variability (MLTV) and heart rate width could suggest reduced adaptability over time.
Decision-makers (e.g., clinicians) might use this to monitor fetuses more closely when these trends appear, even if decelerations (DL, DS, DP) remain low.

2. Feature Prioritization for Modeling
Features with trends (LB, ASTV, MLTV, Width) are more informative for predictive modeling.
Stable features (UC, AC, DS, DP, NSP) might be less useful as predictors and could be given lower weight or excluded for efficiency.

3. EDA Insights for Further Analysis
Temporal trends highlight which features may signal early stress before overt decelerations occur.
Could guide feature engineering, e.g., computing rate of change or moving averages for LB, ASTV, MLTV to capture subtle progression.

4. Monitoring & Alerts
Increasing LB and ASTV, combined with decreasing MLTV/Width, could be used in an early warning system for fetal distress.
Even if NSP remains stable, subtle trends may allow proactive intervention rather than reactive.
