Grade based on marks

```python
number=int(input())
if number>=90:
    print("grade is : A")
elif 75<= number <=89:
    print("grade is : B")
elif 50<=  number <=74:
    print("grade is : C")
else:
    print("failed")
    
```

```python
number=int(input())
if number>0:
    if number%2==0:
        print(number,"is both even and positive")
elif number==0:
    print(number,"the given number is zero")
else:
    print(number,"the given number is negative")
```
