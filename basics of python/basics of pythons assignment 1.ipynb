Hospital Billing Systems

```python
ROOM_RATE_PER_DAY = 2000
DOCTOR_CONSULTATION_FEE = 1500
LAB_TEST_COST = 300
DISCOUNT_THRESHOLD = 10000
DISCOUNT_RATE = 0.10
```

```python
days_stayed = 4
lab_tests_count = 3
medicine_charges = 2400
```

```python
# 1. Room Charges
room_charge = days_stayed * ROOM_RATE_PER_DAY
print(f" Room Charges for {days_stayed} days: ₹{room_charge}")
```

```python
# 2. Lab Test Charges
lab_test_charge = lab_tests_count * LAB_TEST_COST
print(f"2. Lab Test Charges for {lab_tests_count} tests: ₹{lab_test_charge}")
```

```python
# 3. Doctor Consultation Fees
doctor_charge = DOCTOR_CONSULTATION_FEE
print(f"3. Doctor Consultation Fee: ₹{doctor_charge}")
```

```python
# 3. Doctor Consultation Fee Flat
doctor_charge = DOCTOR_CONSULTATION_FEE
print(f"3. Doctor Consultation Fee: ₹{doctor_charge}")
```

```python
# 4. Total Medicine Charges Given
print(f"4. Medicine Charges: ₹{medicine_charges}")
```

```python
# 5. Total Bill Before Discount
total_bill_before_discount = (
    room_charge +
    doctor_charge +
    lab_test_charge +
    medicine_charges
)
```

```python
print(f"\nTotal Bill (Before Discount): ₹{total_bill_before_discount}")
```

```python
# 6. Apply Discount
discount_amount = 0
if total_bill_before_discount > DISCOUNT_THRESHOLD:
    discount_amount = total_bill_before_discount * DISCOUNT_RATE
    print(f"Discount Applied (10%): -₹{discount_amount}")
else:
    print("No discount applied (Total bill is under ₹10,000).")
```

```python
# 7. Final Bill
final_bill = total_bill_before_discount - discount_amount
print(f"\nPatient Final Bill: ₹{final_bill}")
```
