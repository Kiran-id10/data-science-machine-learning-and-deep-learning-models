```python
#here we are removing t5he outliers present in the data
def outlier_capping(df,column):
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    iqr=q3-q1
    lower_extream=q1-1.5*iqr
    upper_extream=q3+1.5*iqr
    df[column]=df[column].apply(lambda x : lower_extream if x<lower_extream else upper_extream if x>upper_extream else x)
for col in df.select_dtypes(["int","float"]).columns:
    outlier_capping(df,col)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(features), columns=features.columns)
vif_data = pd.DataFrame()
vif_data['Feature'] = X_scaled.columns
vif_data['VIF'] = [variance_inflation_factor(X_scaled.values, i)
                   for i in range(X_scaled.shape[1])]

print(vif_data.sort_values(by='VIF', ascending=False))
```

 Variance Inflation Factor (VIF) measures how much the variance of a regression coefficient is inflated due to multicollinearity among predictors.
 Interpretation and Implications

 All VIF values are below 5, which means:
  There is no severe multicollinearity problem.
 All predictors can be safely included in the regression model.

 A VIF > 10 would indicate a serious multicollinearity issue and would require corrective action (e.g., feature removal or dimensionality reduction).
 Linearity – Assumed that the relationship between predictors and the target is linear.

 No perfect multicollinearity – Verified by VIF; assumption holds true.

 Homoscedasticity – Assumed constant variance of residuals.

 Normality of residuals – Assumed residuals follow approximately a normal distribution.
 Since all VIF values < 5 → no multicollinearity issue detected.

 we can proceed with model building using all the above predictors.

1.What is Normalization & Standardization and how is it helpful?

*Normalization scales the values of features to a specific range, typically [0, 1]

*When you know data has a bounded range.
    
*Common in neural networks, k-NN, distance-based algorithms.

Standardization (also called Z-score normalization) transforms data so that it has:
Mean = 0
Standard deviation = 1
Prevents features with large scales from dominating others.

Improves convergence speed in gradient-based models (like neural networks).

Helps models interpret each feature fairly.

Makes distance-based models (like kNN, k-Means) more accurate.

2.What techniques can be used to address multicollinearity in multiple linear regression?

Multicollinearity occurs when two or more independent variables (predictors) in a regression model are highly correlated with each other.
This means one predictor can be linearly predicted from another with high accuracy.

Regression coefficients become unstable and unreliable.

p-values increase — making significant variables appear insignificant.

Model interpretability decreases.

Predictions may still be fine, but the understanding of feature effects becomes poor.
